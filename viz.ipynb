```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("r-1Dh.csv", parse_dates=["timestamp"])

# Count of statuses over time
status_counts = df.groupby(["timestamp", "status"]).size().unstack(fill_value=0)

# Plotting
status_counts.plot(kind='line', marker='o', figsize=(12, 6))
plt.title("Pod Status Over Time")
plt.ylabel("Count")
plt.xlabel("Timestamp")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```
